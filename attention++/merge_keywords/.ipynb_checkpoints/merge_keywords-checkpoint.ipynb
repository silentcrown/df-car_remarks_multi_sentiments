```python
import pandas as pd
```

```python
result = pd.read_csv('../final_result++.csv')
see = pd.read_csv('../see_final_result++.csv')
```

```python
len(result)
```

```python
from collections import Counter
c_ids = list(result['content_id'])
subject = list(result['subject'])
counter_id = Counter(c_ids)
```

```python
mess = see[see['subject'] == '动力']
ind = mess.index.tolist()
change = []
c = 0
ii = []
for i in ind:
    if counter_id[str(mess.loc[i,'content_id'])] == 1:
        text = str(mess.loc[i,'content'])
        if '油' in text:
            change.append('油耗')
            ii.append(i)
        elif '价' in text or '钱' in text:
            change.append('价格')
            ii.append(i)
        elif '胎' in text or '配置' in text or '刹车' in text:
            change.append('配置')
            ii.append(i)
        elif '操控' in text:
            change.append('操控')
            ii.append(i)
        elif '空间' in text:
            change.append('空间')
            ii.append(i)
        elif '安全性' in text:
            change.append('安全性')
            ii.append(i)
        elif '外观' in text:
            change.append('外观')
            ii.append(i)
        elif '内饰' in text or '座椅' in text:
            change.append('内饰')
            ii.append(i)
        elif '动力' in text or '发动机' in text:
            change.append('动力')
            ii.append(i)
        elif '舒' in text:
            change.append('舒适性')
            print('舒适')
            ii.append(i)
        #print(str(mess.loc[i,'content']),see.loc[i,'content'])
        c += 1
print(c)
```

```python
result.loc[ii, 'subject'] =  change
see.loc[ii, 'subject'] = change
result.to_csv('n_result.csv', index = False, encoding = 'utf-8')
see.to_csv('see_n_result.csv', index = False, encoding = 'utf-8')
```
