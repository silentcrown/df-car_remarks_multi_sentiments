```python
import pandas as pd
import re
```

```python
data = pd.read_csv('../../../data/df_sen_sub/train.csv')
data
```

```python
train_line, test_line = [], []
for i in range(len(data)):
    text = str(data.loc[i, 'content'])
    text = re.sub('[，。？ ！:、,?!：…“”''"；;—（）()]', '', text).strip()
    subject = str(data.loc[i, 'subject'])
    sentiment = str(data.loc[i, 'sentiment_value'])
    sentiment_word = str(data.loc[i, 'sentiment_word'])
    if i < 0.9 * len(data):
        for j in range(len(text)):
            #import ipdb;ipdb.set_trace()
            score = '1' if text[j] in sentiment_word else '0' 
            train_line.append(' '.join([text[j], subject, score, sentiment]))
        train_line.append(' ')
    else:
        for j in range(len(text)):
            score = '1' if text[j] in sentiment_word else '0' 
            test_line.append(' '.join([text[j], subject, score, sentiment]))
        test_line.append(' ')   
```

```python
with open('./crf_train_data', 'w') as f:
    f.write('\n'.join(train_line))
with open('./crf_testdata', 'w') as f:
    f.write('\n'.join(test_line))
```
