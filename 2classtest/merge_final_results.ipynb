```python
import pandas as pd
import pickle
```

```python
split_results = {}
subjs = '价格,配置,操控,舒适性,油耗,动力,内饰,安全性,空间,外观'.split(',')
for key in subjs:
    split_results[key] = pickle.load(open('split_result/'+key+'.txt','rb'))
```

```python
test = pd.read_csv('../../../data/df_sen_sub/test_public.csv')
result = pd.read_csv('../../../data/df_sen_sub/submit_example.csv')
result
```

```python
content_ids = []
sentiment_values = []
subjects = []
sentiment_words = []
text = []
for i in range(len(test)):
    c_id = str(test.loc[i, 'content_id'])
    flag = 0
    for key in subjs:
        if c_id in split_results[key]:
            content_ids.append(c_id)
            sentiment_values.append(split_results[key][c_id])
            subjects.append(key)
            sentiment_words.append("")
            text.append(test.loc[i, 'content'])
            flag = 1
    if flag == 0:
        content_ids.append(c_id)
        subjects.append('动力')
        sentiment_values.append(0)
        sentiment_words.append("")   
        text.append(str(test.loc[i, 'content']))
```

```python
real_pred = {"content_id":content_ids, "subject":subjects, "sentiment_value":sentiment_values, "sentiment_word":sentiment_words}
real_df = pd.DataFrame(real_pred)
real_df.to_csv('./final_result.csv', encoding = "utf-8", index = False)
```
